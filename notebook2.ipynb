```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import gaussian_kde

# Charger le dataset (remplacez par le chargement correct de votre dataset)
# df = pd.read_csv('your_dataset.csv')

# Convertir la colonne 'DATE_DIED' en type datetime
df['DATE_DIED'] = pd.to_datetime(df['DATE_DIED'], errors='coerce')

# Créer une colonne pour indiquer si le patient est décédé
df['DECEASED'] = df['DATE_DIED'].notna() & (df['DATE_DIED'] != pd.Timestamp('9999-09-09'))

# Calculer les KDE pour tous les patients, les patients décédés et les patients vivants
kde_all = gaussian_kde(df['AGE'].dropna())
kde_deceased = gaussian_kde(df[df['DECEASED']]['AGE'].dropna())
kde_alive = gaussian_kde(df[~df['DECEASED']]['AGE'].dropna())

# Créer une plage d'âges pour les courbes KDE
age_range = np.linspace(df['AGE'].min(), df['AGE'].max(), 100)

# Calculer les valeurs KDE pour chaque groupe
kde_values_all = kde_all(age_range)
kde_values_deceased = kde_deceased(age_range)
kde_values_alive = kde_alive(age_range)

# Calculer la corrélation de Pearson entre les courbes KDE
correlation_all_deceased = np.corrcoef(kde_values_all, kde_values_deceased)[0, 1]
correlation_all_alive = np.corrcoef(kde_values_all, kde_values_alive)[0, 1]
correlation_deceased_alive = np.corrcoef(kde_values_deceased, kde_values_alive)[0, 1]

# Afficher les corrélations
print(f"Corrélation de Pearson entre KDE tous les patients et patients décédés : {correlation_all_deceased:.2f}")
print(f"Corrélation de Pearson entre KDE tous les patients et patients vivants : {correlation_all_alive:.2f}")
print(f"Corrélation de Pearson entre KDE patients décédés et patients vivants : {correlation_deceased_alive:.2f}")

# Créer une figure pour le graphique
plt.figure(figsize=(12, 6))

# Histogramme pour tous les patients
plt.hist(df['AGE'], bins=30, alpha=0.3, label='Tous les patients', color='blue', density=True)

# Superposer les courbes KDE
plt.plot(age_range, kde_values_all, color='blue', lw=2, label='KDE - Tous les patients')
plt.plot(age_range, kde_values_deceased, color='red', lw=2, label='KDE - Patients décédés')
plt.plot(age_range, kde_values_alive, color='green', lw=2, label='KDE - Patients vivants')

# Histogramme pour les patients décédés
plt.hist(df[df['DECEASED']]['AGE'], bins=30, alpha=0.5, label='Patients décédés', color='red', density=True)

# Histogramme pour les patients vivants
plt.hist(df[~df['DECEASED']]['AGE'], bins=30, alpha=0.5, label='Patients vivants', color='green', density=True)

# Ajouter les labels et le titre
plt.xlabel('Âge')
plt.ylabel('Densité')
plt.title('Distribution des âges avec histogrammes et courbes KDE pour patients vivants et décédés')
plt.legend()

# Afficher le graphique
plt.show()
```

```python
import pandas as pd


def set_to_str(feature_set):
    if len(feature_set) == 1:
        return next(iter(feature_set))
    return str(tuple(sorted(feature_set)))

def search_association_rules(rules, general_terms=None, antecedent_terms=None, consequent_terms=None):
    # Ensure lists are initialized
    general_terms = general_terms or []
    antecedent_terms = antecedent_terms or []
    consequent_terms = consequent_terms or []

    # Collect matching rows
    matching_rows = []
    for idx, row in rules.iterrows():
        antecedent_str = set_to_str(row['antecedents'])
        consequent_str = set_to_str(row['consequents'])

        # Check for matches based on specified terms
        general_match = any(term in antecedent_str or term in consequent_str for term in general_terms)
        antecedent_match = any(term in antecedent_str for term in antecedent_terms)
        consequent_match = any(term in consequent_str for term in consequent_terms)

        # Add row if any match is found
        if general_match or antecedent_match or consequent_match:
            matching_rows.append(row)

    if matching_rows:
        # Create DataFrame for the matching rows and select relevant columns
        df_matches = pd.DataFrame(matching_rows)
        selected_columns = ['antecedents', 'consequents', 'confidence', 'lift', 'support', 'leverage']

        # Display the table with clean formatting, limiting columns for clarity
        pd.set_option('display.max_rows', None)  # Show all rows
        pd.set_option('display.colheader_justify', 'center')  # Center column headers
        pd.set_option('display.width', 120)  # Set width for display
        print("Matching rows:\n", df_matches[selected_columns].to_string(index=False))
    else:
        print("No matches found.")

```

```python
# BINRIESES THE DATA FRAME FOR APRIORI  ALGO
df = pd.read_csv(file_path)

# Function to convert categorical features to binary
def binarize_data(df):
    df['SEX'] = df['SEX'].map({1: 0, 2: 1})  # 0 for returned women, 1 for men

    # Convert AGE to separate binary columns for age groups
    df['AGE_0_19'] = (df['AGE'] < 20).astype(int)
    df['AGE_20_39'] = ((df['AGE'] >= 20) & (df['AGE'] < 40)).astype(int)
    df['AGE_40_59'] = ((df['AGE'] >= 40) & (df['AGE'] < 60)).astype(int)
    df['AGE_60_79'] = ((df['AGE'] >= 60) & (df['AGE'] < 80)).astype(int)
    df['AGE_80_plus'] = (df['AGE'] >= 80).astype(int)

    # Convert CLASIFFICATION_FINAL to binary
    df['CLASIFFICATION_FINAL'] = df['CLASIFFICATION_FINAL'].apply(lambda x: 1 if x < 4 else 0)  # 1 for diagnosed, 0 for not diagnosed

    # Convert PATIENT_TYPE to binary (1 = returned home, 2 = hospitalized)
    df['PATIENT_TYPE'] = df['PATIENT_TYPE'].map({1: 0, 2: 1})  # 0 for returned home, 1 for hospitalized

    # Convert PNEUMONIA to binary
    df['PNEUMONIA'] = df['PNEUMONIA'].apply(lambda x: 1 if x == 1 else 0)  # 1 if yes, 0 if no

    # Convert PREGNANT to binary
    df['PREGNANT'] = df['PREGNANT'].apply(lambda x: 1 if x == 1 else 0)

    # Convert DIABETES to binary
    df['DIABETES'] = df['DIABETES'].apply(lambda x: 1 if x == 1 else 0)

    # Convert COPD to binary
    df['COPD'] = df['COPD'].apply(lambda x: 1 if x == 1 else 0)

    # Convert ASTHMA to binary
    df['ASTHMA'] = df['ASTHMA'].apply(lambda x: 1 if x == 1 else 0)

    # Convert INMSUPR to binary
    df['INMSUPR'] = df['INMSUPR'].apply(lambda x: 1 if x == 1 else 0)

    # Convert HIPERTENSION to binary
    df['HIPERTENSION'] = df['HIPERTENSION'].apply(lambda x: 1 if x == 1 else 0)

    # Convert CARDIOVASCULAR to binary
    df['CARDIOVASCULAR'] = df['CARDIOVASCULAR'].apply(lambda x: 1 if x == 1 else 0)

    # Convert RENAL_CHRONIC to binary
    df['RENAL_CHRONIC'] = df['RENAL_CHRONIC'].apply(lambda x: 1 if x == 1 else 0)

    # Convert OTHER_DISEASE to binary
    df['OTHER_DISEASE'] = df['OTHER_DISEASE'].apply(lambda x: 1 if x == 1 else 0)

    # Convert OBESITY to binary
    df['OBESITY'] = df['OBESITY'].apply(lambda x: 1 if x == 1 else 0)

    # Convert TOBACCO to binary
    df['TOBACCO'] = df['TOBACCO'].apply(lambda x: 1 if x == 1 else 0)

    # Convert USMER to binary: 1 for false, 2 for true
    df['USMER'] = df['USMER'].map({1: 0, 2: 1})  # 0 for not in a medical unit, 1 for in a medical unit

    # One-hot encode the 'MEDICAL_UNIT' column
    med_unit_dummies = pd.get_dummies(df['MEDICAL_UNIT'], prefix='MEDICAL_UNIT')
    df = pd.concat([df, med_unit_dummies], axis=1)

    # Convert INTUBED to binary
    df['INTUBED'] = df['INTUBED'].apply(lambda x: 1 if x == 1 else 0)

    # Convert ICU to binary
    df['ICU'] = df['ICU'].apply(lambda x: 1 if x == 1 else 0)

    # Convert DATE_DIED to binary: 1 for having a death date, 0 for "9999-99-99"
    df['DATE_DIED'] = df['DATE_DIED'].apply(lambda x: 1 if x != "9999-99-99" else 0)

    df.drop(columns=['AGE'], inplace=True)  # Drop the original AGE column
    df.drop(columns=['MEDICAL_UNIT'], inplace=True)  # Drop the original AGE column

    return df







# Apply the binarization function
binarized_df = binarize_data(df)

# Display the first few rows of the binarized dataset
print(binarized_df.head())

# Optionally, save the binarized dataset to a new CSV file
# binarized_df.to_csv("binarized_dataset.csv", index=False)
```

```python


def print_non_binary_values(df):
    # Allowed binary values: 0, 1, True, False
    allowed_values = {0, 1, True, False}

    # Iterate over each column
    for column in df.columns:
        # Check if there are any values not in allowed_values
        non_binary_values = df[~df[column].isin(allowed_values)][column].unique()

        # If non-binary values are found, print the column name and those values
        if len(non_binary_values) > 0:
            print(f"Column '{column}' contains non-binary values: {non_binary_values}")

# Usage example:
print_non_binary_values(binarized_df)

```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Ensure the data is in the right format for Apriori (boolean values)
# Convert all columns that are already binary (0, 1, True, False) to boolean
df_binary = binarized_df.astype(bool)
```

```python

# Apply the Apriori algorithm
# Use a minimum support threshold (e.g., 0.1 means the item is in 10% of transactions)
frequent_itemsets = apriori(df_binary, min_support=0.1, use_colnames=True)

# Generate association rules
rules = association_rules(frequent_itemsets, metric="leverage", min_threshold=-0.1, num_itemsets=[])

# Print the generated rules
print("Association Rules:")
print(rules.shape)
rules.head()

```

```python
import plotly.graph_objects as go
import networkx as nx
import numpy as np

def visualize_directed_graph(rules, threshold=0.25, threshold_metric='confidence'):
    # Ensure the threshold metric exists in the DataFrame
    if threshold_metric not in rules.columns:
        raise ValueError(f"The threshold metric '{threshold_metric}' is not a column in the rules DataFrame.")

    # Create directed graph
    G = nx.DiGraph()

    # Add edges from rules based on the specified threshold metric
    for _, row in rules.iterrows():
        antecedent_str = set_to_str(row['antecedents'])
        consequent_str = set_to_str(row['consequents'])
        metric_value = row[threshold_metric]
        if metric_value >= threshold:
            G.add_edge(antecedent_str, consequent_str, weight=metric_value)

    if len(G.edges()) == 0:
        print("No edges meet the threshold criteria")
        return

    # Calculate layout using Fruchterman-Reingold force-directed algorithm
    pos = nx.spring_layout(G, k=3, iterations=500)

    # Create separate edge traces for each edge to handle different colors
    edge_traces = []
    arrow_x = []
    arrow_y = []

    # Get all metric values for color scaling
    metric_values = [G.edges[edge]['weight'] for edge in G.edges()]

    # Normalize without logarithmic scaling
    min_metric = min(metric_values)
    max_metric = max(metric_values)

    for edge in G.edges(data=True):
        x0, y0 = pos[edge[0]]
        x1, y1 = pos[edge[1]]

        # Directly normalize without log scaling
        normalized_value = (edge[2]['weight'] - min_metric) / (max_metric - min_metric) if max_metric != min_metric else 0.5
        color = f'rgb({int(255*normalized_value)}, 0, {int(255*(1-normalized_value))})'

        # Create edge trace
        edge_trace = go.Scatter(
            x=[x0, x1, None],
            y=[y0, y1, None],
            line=dict(width=2, color=color),
            hoverinfo='text',
            text=f"{threshold_metric.capitalize()}: {edge[2]['weight']:.2f}",
            mode='lines'
        )
        edge_traces.append(edge_trace)

        # Add arrow at the end of the edge
        arrow_x.append(x1)
        arrow_y.append(y1)

    # Create figure
    fig = go.Figure()

    # Add all edge traces
    for edge_trace in edge_traces:
        fig.add_trace(edge_trace)

    # Add arrows
    fig.add_trace(go.Scatter(
        x=arrow_x, y=arrow_y,
        mode='markers',
        marker=dict(
            size=8,
            color='#888',
            symbol='triangle-right'
        ),
        hoverinfo='none',
        name='Arrows'
    ))

    # Add nodes
    node_x = []
    node_y = []
    node_text = []
    for node in G.nodes():
        x, y = pos[node]
        node_x.append(x)
        node_y.append(y)
        formatted_node = str(node).replace("'", "").replace("(", "").replace(")", "")
        node_text.append(formatted_node)

    fig.add_trace(go.Scatter(
        x=node_x, y=node_y,
        mode='markers+text',
        hoverinfo='text',
        text=node_text,
        textposition="top center",
        marker=dict(
            size=30,
            color='lightblue',
            line=dict(width=2, color='darkblue')
        ),
        name='Nodes'
    ))

    # Add color scale
    fig.add_trace(go.Scatter(
        x=[None],
        y=[None],
        mode='markers',
        marker=dict(
            colorscale=[[0, 'blue'], [1, 'red']],
            showscale=True,
            cmin=min_metric,
            cmax=max_metric,
            colorbar=dict(
                title=threshold_metric.capitalize(),
                thickness=15,
                len=0.5,
                y=0.5
            )
        ),
        hoverinfo='none',
        showlegend=False
    ))

    # Update layout
    fig.update_layout(
        title=f'Association Rules Network ({threshold_metric.capitalize()} ≥ {threshold:.2f})',
        showlegend=False,
        hovermode='closest',
        margin=dict(b=20, l=5, r=5, t=40),
        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
        width=1000,
        height=800,
        plot_bgcolor='white'
    )

    return fig
```

```python
import pandas as pd

def calculate_survival_rate(df):
    # Ensure necessary columns are present
    if 'CLASIFFICATION_FINAL' not in df.columns or 'SEX' not in df.columns or 'DECEASED' not in df.columns:
        raise ValueError("The dataframe must contain 'CLASIFFICATION_FINAL', 'SEX', and 'DECEASED' columns.")

    # If there's no 'DECEASED' column, ensure there's a column that marks patient survival (e.g., 'SURVIVAL_STATUS')
    # Assume 0 means alive, and 1 means deceased, if necessary adjust column names

    # Survival rate for the general population by sex
    survival_by_sex = df.groupby('SEX').apply(
        lambda x: (x['DECEASED'] == 0).sum() / len(x)  # Assuming 0 means survived, 1 means deceased
    ).reset_index(name='survival_rate')
    survival_by_sex['SEX'] = survival_by_sex['SEX'].map({0: 'Female', 1: 'Male'})

    # Survival rate for COVID diagnosed patients by sex
    covid_df = df[df['CLASIFFICATION_FINAL'] == 1]  # Filter for patients diagnosed with COVID
    survival_covid_by_sex = covid_df.groupby('SEX').apply(
        lambda x: (x['DECEASED'] == 0).sum() / len(x)  # Assuming 0 means survived, 1 means deceased
    ).reset_index(name='survival_rate')
    survival_covid_by_sex['SEX'] = survival_covid_by_sex['SEX'].map({0: 'Female', 1: 'Male'})

    # Calculate the difference in survival rate between male and female for the general population
    general_female_rate = survival_by_sex[survival_by_sex['SEX'] == 'Female']['survival_rate'].values[0]
    general_male_rate = survival_by_sex[survival_by_sex['SEX'] == 'Male']['survival_rate'].values[0]
    survival_difference_general = general_male_rate - general_female_rate

    # Calculate the difference in survival rate between male and female for COVID diagnosed patients
    covid_female_rate = survival_covid_by_sex[survival_covid_by_sex['SEX'] == 'Female']['survival_rate'].values[0]
    covid_male_rate = survival_covid_by_sex[survival_covid_by_sex['SEX'] == 'Male']['survival_rate'].values[0]
    survival_difference_covid = covid_male_rate - covid_female_rate

    # Return the survival rates and the difference
    return survival_by_sex, survival_covid_by_sex, survival_difference_general, survival_difference_covid

# Example usage
survival_general, survival_covid, difference_general, difference_covid = calculate_survival_rate(df)

print("Survival Rate for General Population by Sex:")
print(survival_general)

print("\nSurvival Rate for COVID Diagnosed Patients by Sex:")
print(survival_covid)

print(f"\nDifference in Survival Rate between Males and Females for General Population: {difference_general*100:.2f}%")
print(f"Difference in Survival Rate between Males and Females for COVID Diagnosed Patients: {difference_covid*100:.2f}%")
```

```python
import pandas as pd

def calculate_average_age_by_sex(df):
    # Ensure 'AGE' and 'SEX' columns are present
    if 'AGE' not in df.columns or 'SEX' not in df.columns:
        raise ValueError("The dataframe must contain 'AGE' and 'SEX' columns.")

    # Calculate the average age for men and women
    average_age_by_sex = df.groupby('SEX')['AGE'].mean().reset_index(name='average_age')
    average_age_by_sex['SEX'] = average_age_by_sex['SEX'].map({0: 'Female', 1: 'Male'})

    return average_age_by_sex

# Example usage
average_age = calculate_average_age_by_sex(df)

print("Average Age by Sex:")
print(average_age)
```

```python
def calculate_sex_percentage(sex_ratio):
    # Calculate the total population factor (1 female + sex_ratio males)
    total_population_factor = 1 + sex_ratio

    # Calculate the percentage of females and males
    percentage_females = 1 / total_population_factor * 100
    percentage_males = sex_ratio / total_population_factor * 100

    return percentage_females, percentage_males

# Example usage with a sex ratio of 0.96 males per female
sex_ratio = 0.96 # from https://en.wikipedia.org/wiki/Demographics_of_Mexico
females, males = calculate_sex_percentage(sex_ratio)

print(f"Percentage of females: {females:.2f}%")
print(f"Percentage of males: {males:.2f}%")
```
